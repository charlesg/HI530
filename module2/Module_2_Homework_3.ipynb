## Question 1

For the first graph, use source as ORIGIN and target as CARRIER. What can you infer from this graph?

```python
import numpy as np
import networkx as nx
import pandas as pd
import matplotlib.pylab as plt
```

```python
fd_df = pd.read_csv('FlightDelays.csv')
fd_df
```

```python
G = nx.from_pandas_edgelist(fd_df, source='ORIGIN', target='CARRIER')
 
isCarrier = [n in set(fd_df.CARRIER) for n in G.nodes()]
pos = nx.spring_layout(G, k=0.13, iterations=60, scale=0.5)
plt.figure(figsize=(10,10))
nx.draw_networkx(G, pos=pos, with_labels=True,
                 edge_color='lightgray',
                 node_color=['gray' if carrier else 'black' for carrier in isCarrier],
                 node_size=[50 if carrier else 200 for carrier in isCarrier])
plt.axis('off')
plt.show()
```

### Interpretation

One airport (DCA), seems to have a lot more carriers than the other two airports (BWI and IAD). Two carriers (DH and RU) connect DCA to IAD while a single carrier, RU, connects DCA to BWI.
